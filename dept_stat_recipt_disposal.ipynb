### Finding Efficiency of each Department/State/Ministry ###

```python
df['Efficiency']=(df['Total Disposal (01.01.2016 to 01.11.2019)']/df['Total Receipts (01.01.2016 to 01.11.2019)'])*100
```

```python
df_eff = df.sort_values(by='Efficiency',ascending=False)
```
